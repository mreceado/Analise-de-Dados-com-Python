### Prática Função loadtxt()

A função numpy.loadtxt() carrega dados de um ficheiro de texto e fornece uma abordagem rápida para ficheiros de texto simples.

dtype	Tipo de dados do array resultante 
comments	Caracteres ou lista de caracteres utilizados para indicar o início de um comentário
delimiter	Delimitador a utilizar para analisar o conteúdo do ficheiro txt.
converters	Número de coluna de mapeamento do dicionário para uma função que irá analisar a string de colunas para o valor desejado.
skiprows	Que fila/fileiras saltar
usecols	Os índices das colunas a serem lidos
unpack	Transpor a matriz devolvida, para que os argumentos possam ser desembalados utilizando x, y, z = loadtxt(...). [unpack=True].
ndim	Número mínimo de dimensões no array devolvida
encoding	Codificação utilizada para descodificar o ficheiro de entrada.
max_rows	Número máximo de linhas a ler após as linhas skiprows.

```python
import numpy as np
from io import StringIO
```

```python
f = StringIO("3 6 8 \n12 9 1 \n2 3 4")
```

```python
a = np.loadtxt(f)
```

```python
# A função loadtxt por defeito apresenta o conteudo como sendo float 
print(a)
```

```python
# Converte float para inteiros
b = np.array(a, dtype = "int64")
```

```python
print(b)
```

```python
f = StringIO("3, 6, 8 \n12, 9, 1 \n 2, 3, 4")
a = np.loadtxt(f,dtype="int", delimiter = ",")

print(a)
```

```python
f = StringIO("3 6 8 \n7 9 1 \n2 3 4")
(x,y,z) = np.loadtxt(f, dtype = "int", unpack = True)
```

```python
print(x)
print(y)
print(z)
```

```python
# Imprimir a coluna 0 e 1 com o uso do delimiter
f = StringIO("3, 6, 8 \n12, 9, 1 \n 2, 3, 4")
a = np.loadtxt(f,dtype="int", usecols = (0,1), delimiter = ",")

print(a)
```

```python

f = StringIO("3 6 8 \n12 9 1 \n 2 3 4")
a = np.loadtxt(f,dtype="int", skiprows = 0)

print(a)
```

```python
f = StringIO("3 6 8 \n12 9 1 \n 2 3 4")
b = np.loadtxt(f, dtype = "int", usecols = (1,2), skiprows = 1)

print(b)
```

```python
f = StringIO("3 6 8 \n12 9 1 \n 2 3 4")
c = np.loadtxt(f, dtype = "int", skiprows = 2)

print(c)
```

### Matplotlib 2D

O Matplotlib é uma biblioteca para a visualização de dados em Python. Ele apresenta uma API orientada a objetos que permite a criação de gráficos em 2D de uma forma simples e com poucos comandos. A ferramenta disponibiliza diversos tipos de gráficos, como em barra, em linha, em pizza, histogramas entre outras opções. A API foi projetada para ser compatível com o MATLAB que, apesar de ser referência na área de processamento numérico, não é opensource.

#### Visualização de dados com PyPlot

O PyPlot é um módulo do matplotlib para criação de gráficos. Para utilizá-lo é necessário fazer a importação:

```python
import matplotlib.pyplot as plt
```

```python
# Criando um gráfico
plt.plot([1,3,5], [2,5,7])
plt.show()
```

```python
# Definindo variáveis
x = [1,3,5]
y = [1,2,5]

# Criando um gráfico
plt.plot(x,y)

# Atribuindo um título ao gráfico
plt.title('Exemplo utilizando Plot')
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')

# Atribuindo uma legenda
plt.plot(x, y, label = 'Uma legenda')
plt.legend()

# Exibindo o gráfico gerado
plt.show()
```

#### Gráfico de Barras

```python
# Definindo as variáveis
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]

x2 = [1, 3, 5, 7, 9]
y2 = [7, 8, 2, 4, 2]

plt.bar(x, y, label = 'Barrar1', color = 'r')
plt.bar(x2, y2, label = 'Barrar2', color = 'y')

plt.legend()

plt.show()
```

#### Pie Chart

```python
# Definindo as variáveis
fatias = [6, 4, 8]
actividades = ['X', 'Y', 'Z']
colunas = ['r', 'm', 'y']

# Criando um gráfico
plt.pie(fatias, labels = actividades, colors = colunas, startangle=90, shadow = True, explode = (0.1, 0, 0))
plt.show()
```

#### Scatterplot

```python
# Definindo variáveis 
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8]

# Criando um gráfico
plt.scatter(x, y, label = 'Pontos', color = 'c', marker = '*', s = 100)
plt.legend()
plt.show()
```

Muitas vezes, gostaríamos de um gráfico de superfície em vez de um gráfico de linha ao plotar em três dimensões. Na plotagem de superfícies tridimensionais, desejamos fazer um gráfico de alguma relação f (x, y) . Na plotagem de superfície, todos os pares (x,y) devem ser fornecidos. Isso não é simples de fazer usando vetores. Portanto, na plotagem de superfície, a primeira estrutura de dados que você deve criar é chamada de malha. Dadas listas/arrays de valores x e y, uma malha é uma listagem de todas as combinações possíveis de x e y. Em Python, a malha é dada como duas matrizes X e Y onde X (i,j) e Y (i,j) definem possíveis pares (x,y). Uma terceira matriz, Z, pode então ser criada de forma que Z (i,j) = f (X (i,j), Y (i,j)). Uma malha pode ser criada usando a função np.meshgrid em Python. a grade de malhaA função tem as entradas x e y são listas contendo o conjunto de dados independente. As variáveis de saída X e Y são conforme descrito anteriormente.

TENTE! Crie uma malha para x = [1, 2, 3, 4] e y = [3, 4, 5] usando a função meshgrid .

```python
x = [1, 2, 3, 4]
y = [3, 4, 5]

X, Y = np.meshgrid(x,y)
```

TENTE! Faça um gráfico da superfícief(x,y)=sin(x)⋅cos(y)for−5≤x≤5,−5≤y≤5
usando a função plot_surface . Tome cuidado para usar uma discretização suficientemente fina em x e y para fazer o gráfico parecer suave.

```python
fig = plt.figure(figsize = (12, 10))
ax = plt.axes(projection = '3d')

x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X,Y = np.meshgrid(x,y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```

Você notará que o gráfico de superfície mostra cores diferentes para diferentes elevações, amarelo para mais alto e azul para mais baixo, pois usamos o mapa de cores plt.cm.cividis no gráfico de superfície. Você pode mudar para diferentes esquemas de cores para a plotagem da superfície. Estes são deixados como exercícios. Também plotamos uma barra de cores para mostrar as cores correspondentes a diferentes valores.

Poderíamos ter subtramas de diferentes tramas 3D também. Poderíamos usar a função add_subplot do objeto figure que criamos para gerar as subtramas para casos 3D.

' EXPERIMENTE! Faça um subplot de 1 por 2 para plotar os dados X, Y, Z acima no gráfico de estrutura de arame e no gráfico de superfície.

```python
fig = plt.figure(figsize = (12,6))

ax = fig.add_subplot(1, 2, 1, projection = '3d')
ax.plot_wireframe(X,Y,Z)
ax.set_title('Wireframe Plot')

ax = fig.add_subplot(1, 2, 2, projection = '3d')
ax.plot_surface(X,Y,Z)
ax.set_title('Surface Plot')

plt.tight_layout()

plt.show()
```
