### Criando Gráficos com Seaborn

```python
# Carregando um dos datasets que vem com o seaborn
dados = sea.load_dataset("tips")
```

```python
dados.head()
```

```python
# O método jointplot cria plot de 2 variáveis com gráficos bivariados e univariados
sea.jointplot(dados, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
# O método mlplot() cria plot com dados e modelos de regressão
# Constroi um gráfico, usando como base de dados o arquivo "dados", cujo objectivo é analisar o "total_bill" e o "tip", em 
# função da coluna "smoker"
sea.lmplot(data = dados, x = "total_bill", y = "tip", col = "smoker")
```

```python
# Construindo um dataframe com pandas
df = pd.DataFrame()
```

```python
df.shape
```

```python
df.head()
```

```python
# lmplot
sea.lmplot(data = df, x = 'idade', y = 'peso', fit_reg = 'True')
```

```python
# kdeplot
sea.kdeplot(df.idade)
```

```python
# kdeplot
sea.kdeplot(df.peso)
```

```python
# distplot
sea.displot(df.peso)
```

```python
# Histograma
plt.hist(df.idade, alpha = .3)
sea.rugplot(df.idade)
```

```python
# Box plot
sea.boxplot(df.idade, color = 'm')
```

```python
# Box plot
sea.boxplot(df.peso, color = 'y')
```

```python
# Violin plot
sea.violinplot(df.idade, color = 'g')
```

```python
# Violin plot
sea.violinplot(df.peso, color = 'c')
```

```python
# Clustermap
sea.clustermap(df)
```

### Usando Matplotlib, Seaborn, Numpy e Pandas na Criação de Gráfico Estatístico

```python
# Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variáveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})
                      
# Cria os dados para a variável flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'Não Fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Style
sea.set(style = "ticks")

# lmplot
sea.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumante',
           palette = ['tab:blue', 'tab:orange'],
           height = 7)

# Labels e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação entre Altura e Peso de Fumantes e Não Fumantes')

# Remove as bordas
sea.despine()

# Show
plt.show()
```
