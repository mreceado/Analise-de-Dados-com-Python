```python
# Agrupar os dados por uma determinada coluna
df.groupby("Regiao")["Pais"].nunique()
```

```python
# A Quantidade por Segmento
df.groupby('Segmento')['Quantidade'].sum()
```

```python
# A m√©dia de qtd por segmento
df.groupby('Segmento')['Quantidade'].mean()
```

```python
# Multiplica√ß√£o de colunas
df['Receita'] = df['Valor_Venda'].mul(df['Quantidade'])
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Moda - valor que mais aparece
moda_Qtd = df['Quantidade'].value_counts().index[0]
```

```python
# Preenchendo os valores ausentes em quantidade
df['Quantidade'].fillna(value=moda_Qtd, inplace = True)
```

```python
# Percorre a coluna Receita e encontra  o valor que mais aparece
moda_Rec = df['Receita'].value_counts().index[0]
```

```python
# preenche os valores ausentes da coluna Receita
df['Receita'].fillna(value = moda_Rec, inplace = True)
```

```python
# Convertendo data_pedido em datetime
df['Data_Pedido'] = pd.to_datetime(df['Data_Pedido'], dayfirst= True)
```

```python
df.info()
```

```python
# Mostrar o resumo das receitas por ano
df.groupby(df['Data_Pedido'].dt.year)['Valor_Venda'].sum()
```

```python
# Plot
df.groupby(df['Data_Pedido'].dt.year)['Valor_Venda'].sum().plot.bar();
```

```python
# Criar uma coluna com o ano
df['Ano'] = df['Data_Pedido'].dt.year
```

```python
df.head()
```

```python
# Criar as colunas M√™s e Dia
df['Mes'], df['Dia'] = (df['Data_Pedido'].dt.month, df['Data_Pedido'].dt.day)
```

```python
# Qual √© a data mais antiga do nosso dataset
data_mais_antiga = df['Data_Pedido'].min()
data_mais_antiga
```
