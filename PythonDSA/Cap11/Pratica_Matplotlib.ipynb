### Gráfico de Área Empilhada

Stack plots, também conhecidos como gráficos de área empilhada, são um tipo de plotagem usados para visualizar a mudança relativa de diversas variáveis ao longo do tempo. Eles consistem em várias áreas coloridas empilhadas umas sobre as outras, onde a altura de cada área representa a magnitude da variável correspondente e a largura representa a escala de tempo.

Os stack plots são úteis para mostrar como as partes contribuem para o todo ao longo do tempo. Por exemplo, eles podem ser usados para visualizar a mudança relativa de diferentes setores de uma indústria ao longo do tempo, ou a distribuição relativa de receitas e despesas de uma empresa em um determinado período.

```python
dias = [1, 2, 3, 4, 5]
dormir = [7, 8, 6, 77, 7]
comer = [2, 3, 4, 5, 3]
trabalhar = [7, 8, 7, 2, 2]
passear = [8, 5, 7, 8, 13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors = ['m', 'c', 'r', 'k', 'b'])
plt.show()
```

```python
valor = [12, 15, 20, 17, 8]
variavel = ['Criança', 'Adolescente', 'Jovem', 'Adulto', 'Idoso']
cores = ['olive', 'lime', 'violet', 'royalblue', 'red']
```

```python
plt.pie(valor, labels = variavel, colors = cores, startangle = 90, shadow = True)
plt.show()
```

```python
# Gráfico de Linha

# Linspace retorna números espaçados de modo uniforme em um intervalo. Dessa forma, dado um ponto inicial e final, assim 
# como a quantidade de valores, linspace irá distribuí-los uniformemente em uma matriz NumPy.
# linspace inclui o último valor, range e arange excluem o último valor

# Dados
x = linspace(0, 5, 10)
y = x**2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0, 0, 0.8, 0.8])

# Cria o plot
axes.plot(x, y, 'r')

# Labels e título
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Gráfico de Linha')
```

```python
# Gráficos de linha com 2 figuras

# Dados
x = linspace(0, 5, 10)
y = x**2

# Cria a figura
fig = plt.figure()

# Cria os eixos
eixo1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # Eixos da figura principal
eixo2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # Eixos da figura secundária

# Desenha e Legenda a figura principal
eixo1.plot(x, y, 'r')
eixo1.set_xlabel('X')
eixo1.set_ylabel('Y')
eixo1.set_title('Figura Principal')

# Figura secundária
eixo2.plot(y, x, 'g')
eixo2.set_xlabel('y')
eixo2.set_ylabel('x')
eixo2.set_title('Figura Secundária');  # O uso do ";" retira a mensagem "Text(0.5, 1.0, 'Figura Secundária')" no final, 
                                       # aquando da exibição do gráfico
```

```python
# Gráficos de linha em paralelo

# Dados
x = linspace(0, 5, 10)
y = x**2

# Divide a área de plotagem em 2 subplots
fig, eixos = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot / Ao invés de desenhar o eixo1 e o eixo2 em separado, ele usa um loop para o fazer
for ex in eixos:
    ex.plot(x, y, 'r')
    ex.set_xlabel('X')
    ex.set_ylabel('Y')
    ex.set_title('Título')
    
# Ajusta o layout
fig.tight_layout()
```

```python
# Gráficos de linhas com diferentes escalas

# Dados
x = linspace(0, 5, 10)
y = x**2

# Cria os subplots
fig, eixos = plt.subplots(1, 2, figsize = (10, 4))

# Cria o plot1
eixos[0].plot(x, x**2, x, exp(x))
eixos[0].set_title("Escala Padrão")

# Cria o plot2
eixos[1].plot(x, x**2, x, exp(x))
eixos[1].set_yscale("log")
eixos[1].set_title("Escala Logaritmica");
```

```python
# Grid (grade)
eixos[0].plot(x, x**2, x, x**3, lw = 2)
eixos[0].grid(True)

# Dados
x = linspace(0, 5, 10)
y = x**2

# Cria os subplots
fig, eixos = plt.subplots(1, 2, figsize = (10, 3))

# Grid Padrão
eixos[0].plot(x, x**2, x, x**3, lw = 2)
eixos[0].grid(True)

# Grid Customizado
eixos[1].plot(x, x**2, x, x**3, lw = 2)
eixos[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```

```python
# Histogramas

# Dados
n = np.random.randn(100000)

# Cria os subplots
fig, eixos = plt.subplots(1, 2, figsize = (12, 4))

# Plot 1
eixos[0].hist(n)
eixos[0].set_title("Histograma Padrão")
eixos[0].set_xlim((min(n), max(n)))

# Plot 2
eixos[1].hist(n, cumulative = True, bins = 50)
eixos[1].set_title("Histograma Cumulativo")
eixos[1].set_xlim((min(n), max(n)));
```

```python
# Dados

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

# Função para um mapa de cores
def ColorMap(phi_m, phi_p):
    return (+ alpha - 2 * np.cos(phi_p)*cos(phi_m) + alpha * np.cos(phi_ext - 2*phi_p))

# Mais dados
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T
```

```python
# Cria a Figura

fig = plt.figure(figsize = (14, 6))

# Adiciona o subplot 1 com projecção 3d
eixo = fig.add_subplot(1, 2, 1, projection = '3d')
p = eixo.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# Adiciona o subplot 2 com projecção 3d
eixo = fig.add_subplot(1, 2, 2, projection = '3d')
p = eixo.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Cria a barra de cores como legenda
cb = fig.colorbar(p, shrink=0.5)
```
